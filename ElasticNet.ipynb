Elastic NET

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer
import csv

def get_elasticnet(wq, df):
    # train model
    df_clean = df.dropna(subset=[wq])
    X = df_clean[climate_vars + urban_forms]
    y = df_clean[[wq]]
    
    # impute the missing values in X
    imputer = SimpleImputer(strategy='mean') 
    X_imputed = imputer.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    model = ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=42)  # You can tune alpha and l1_ratio
    model.fit(X_train_scaled, y_train)

    # predict
    y_pred = model.predict(X_test_scaled)
    mse = mean_squared_error(y_test, y_pred)
    print(f'Mean Squared Error for {wq}: {mse}')
    
    coef = model.coef_
    feature_names = X.columns
    # for feature, coeff in zip(feature_names, coef):
    #     print(f'{feature}: {coeff}')
    features_with_coef = list(zip(feature_names, coef))
    features_with_coef_sorted = sorted(features_with_coef, key=lambda x: x[1], reverse=True)
    # print(features_with_coef_sorted)

    # write to csv
    output_file = f'../output/elasticNet/{wq}_elasticNet.csv'

    with open(output_file, mode='w', newline='') as file:
        writer = csv.writer(file)
        # Write header
        writer.writerow(['Feature', 'Coefficient'])
        
        # Write sorted data
        writer.writerows(features_with_coef_sorted)

for wq in wq_i:
    model = get_elasticnet(wq, df)
```

```python
y_pred = model.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

coef = model.coef_
feature_names = X.columns
for feature, coeff in zip(feature_names, coef):
    print(f'{feature}: {coeff}')
```
