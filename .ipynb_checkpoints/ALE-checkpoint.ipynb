```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
import seaborn as sns

from sklearn.ensemble import RandomForestRegressor # ??? 
```

```python
# for ale 1d plots
# https://christophm.github.io/interpretable-ml-book/ale.html
# https://towardsdatascience.com/deep-dive-on-accumulated-local-effect-plots-ales-with-python-0fc9698ed0ee
# https://docs.seldon.io/projects/alibi/en/stable/index.html

from alibi.explainers import ALE, plot_ale
```

```python
df = pd.read_csv('../data/processed/WQ_BE_CLM_By_Month.csv')
df
```

```python
features = ['tmean', 'ppt', 'ppt_rx1', 'RoadDensity',
           'LaneDensity', 'GravelDensity', 'AsphaltDensity', 'ConcreteDensity',
           'StateRdDensity', 'CountyRdDensity', 'CityRdDensity', 'ParcelCtDensity',
           'ParcelAvgArea', 'AgPercent', 'CommercialPercent', 'GreenPercent',
           'IndustryPercent', 'ServicePercent', 'ResidentialPercent',
           'VaccantPercent', 'OpenDevPercent', 'LowDevPercent', 'MedDevPercent',
           'HighDevPercent', 'MedianFloors', 'MedianBuiltYr', 'RangeBuiltYr',
           'FootprintDensity', 'UnitDensity', 'SoilBPercent', 'SoilCPercent',
           'SoilDPercent', 'SoilDepth', 'Slope']

def get_Xy_model(df, var, features=features):
    df_filtered = df.dropna(subset=[var] + features)

    X = df_filtered[features]
    y = df_filtered[var]

    model = RandomForestRegressor(random_state=123)
    model.fit(X.to_numpy(), y)
    return X, y, model

def plot_1D_ale(df, var, sharey):
    '''
    sharey – A parameter specifying whether the y-axis of the ALE curves should be on the same scale for several features. 
            Possible values are: 'all' | 'row' | None.

    '''
    X, y, model = get_Xy_model(df, var)
    ale = ALE(model.predict , feature_names=X.columns, target_names=['tmean'])
    exp = ale.explain(X.to_numpy())

    plot_ale(exp, features='all', targets='all',sharey=sharey,n_cols=4,fig_kw={'figwidth':18, 'figheight': 30})
```

```python
plot_1D_ale(df,'Discharge',sharey='all')
```

```python
plot_1D_ale(df,'Discharge',sharey='row')
```

```python
import warnings
warnings.filterwarnings('ignore')
from PyALE import ale

def get_df(var):
    irf = pd.read_csv(f'../data/processed/intermediate/irf/irf_interaction_by_month_{var}.csv')
    return irf
    
def get_interact2(var):
    irf = get_df(var)
    filtered_irf = irf[(irf['cpe']>0.5) & (irf['stability']>0.5)]
#     var_interactions = filtered_irf['int']
#     cpe = filtered_irf['cpe']
#     stability = filtered_irf['stability']

    # TP (irf['cpe']>0.5) & (irf['stability']>0.5)] 22
    # TSS: (irf['cpe']>0.5) & (irf['stability']>0.5)] 5
    # ecol: 4
    interact2 = []
    for index,row in filtered_irf.iterrows():
        iteract = row['int']
        cpe = row['cpe']
        stability = row['stability']
        iteract = iteract.replace('+','').replace('-','')
        names = iteract.split('_')
        if len(names) == 2:
            interact2.append({'names':names,'cpe':cpe,'stability':stability})
    return interact2

def plot_ale_interaction(df, var):
    print('============== ', var, ' ==============')
    interact2res = get_interact2(var)
    X, y, model = get_Xy_model(df, var)
    
    for interact2 in interact2res:
        name = interact2['names']
        print('name: ', name)
        print('cpe: ', interact2['cpe'])
        print('stability: ', interact2['stability'])
        ale_eff = ale(X=X, model=model, feature=[name[0],name[1]], grid_size=100)
        
```
